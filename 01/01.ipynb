```python
#rename columns (get rid of French title)
df.columns = ['Electoral District Number','Electoral District Name_English','Electoral District Name_French','Polling Station Number',
             'Polling Station Name','Void Poll Indicator','No Poll Held Indicator','Merge With','Rejected Ballots for Polling Station',
             'Electors for Polling Station','Candidate’s Family Name','Candidate’s Middle Name','Candidate’s First Name',
              'Political Affiliation Name_English','Political Affiliation Name_French','Incumbent Indicator',
              'Elected Candidate Indicator','Candidate Poll Votes Count']
```

```python
df.dtypes
```

```python
df.loc[df['Polling Station Number'] == ' 60A', 'Polling Station Number'] = '6001'
df.loc[df['Polling Station Number'] == ' 60B', 'Polling Station Number'] = '6002'
df.loc[df['Polling Station Number'] == ' 79A', 'Polling Station Number'] = '7901'
df.loc[df['Polling Station Number'] == ' 79B', 'Polling Station Number'] = '7902'
df.loc[df['Polling Station Number'] == ' 96A', 'Polling Station Number'] = '9601'
df.loc[df['Polling Station Number'] == ' 96B', 'Polling Station Number'] = '9602'
df.loc[df['Polling Station Number'] == ' 107A', 'Polling Station Number'] = '10701'
df.loc[df['Polling Station Number'] == ' 107B', 'Polling Station Number'] = '10702'
df.loc[df['Polling Station Number'] == ' 137A', 'Polling Station Number'] = '13701'
df.loc[df['Polling Station Number'] == ' 137B', 'Polling Station Number'] = '13702'
df.loc[df['Polling Station Number'] == ' S/R 1', 'Polling Station Number'] = '100001'
df.loc[df['Polling Station Number'] == ' S/R 2', 'Polling Station Number'] = '100002'
df[df['Polling Station Number']=='100002']
```

```python
df = df.astype({"Polling Station Number": int})
```

```python
#create a pivot table
df_pivot = pd.pivot_table(df,index = ['Polling Station Number'], 
                          columns = ['Political Affiliation Name_English'], values = ['Candidate Poll Votes Count'])
```

```python
df_pivot
```

```python
#the result has multi-index
df_pivot['Candidate Poll Votes Count'].columns
```

```python
df_votes = df_pivot['Candidate Poll Votes Count']
df_votes.columns.values
```

```python
df_votes['Rejected Ballots for Polling Station'] = df.groupby('Polling Station Number')[['Rejected Ballots for Polling Station']].sum()
```

```python
df_votes
```

```python
df_votes.dtypes
```

```python
df_votes['Grand Total'] = df_votes.apply(lambda x: x.sum(), axis=1)
```

```python
df_votes
```

```python
df_votes['Electors for Polling Station'] = df.groupby('Polling Station Number')[['Electors for Polling Station']].mean()
```
