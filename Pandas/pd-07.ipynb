### 61. Compute maximum possible absolute correlation value of each column against other columns in df.

```python
df = pd.DataFrame(np.random.randint(1,100, 80).reshape(8, -1), columns=list('pqrstuvwxy'), index=list('abcdefgh'))
df
```

```python
# Solution
abs_corrmat = np.abs(df.corr())
abs_corrmat
```

```python
#apply(): default apply to column
max_corr = abs_corrmat.apply(lambda x: sorted(x)[-2])
max_corr
```

```python
np.round(max_corr.tolist(), 2)
```

### 62. Compute the minimum-by-maximum for every row of df.

```python
df = pd.DataFrame(np.random.randint(1,100, 80).reshape(8, -1))
df
```

```python
# Solution 1
min_by_max = df.apply(lambda x: np.min(x)/np.max(x), axis=1)
min_by_max
```

```python
# Solution 2
min_by_max = np.min(df, axis=1)/np.max(df, axis=1)
min_by_max
```

### 63. Create a new column 'penultimate' which has the second largest value of each row of df.

```python
df = pd.DataFrame(np.random.randint(1,100, 80).reshape(8, -1))
df
```

```python
out = df.apply(lambda x: x.sort_values().unique()[-2], axis=1)
df['penultimate'] = out
df
```

### 64.  1. Normalize all columns of df by subtracting the column mean and divide by standard deviation.   2. Range all columns of df such that the minimum value in each column is 0 and max is 1.

```python
df = pd.DataFrame(np.random.randint(1,100, 80).reshape(8, -1))
df
```

```python
# Solution Q1
out1 = df.apply(lambda x: ((x - x.mean())/x.std()).round(2))
out1
```

```python
# Solution Q2
out2 = df.apply(lambda x: ((x.max() - x)/(x.max() - x.min())).round(2))
out2 
```

### 66. Replace both values in both diagonals of df with 0.

```python
df = pd.DataFrame(np.random.randint(1,100, 100).reshape(10, -1))
df
```

```python
# Solution
# difference between loc, iloc, at, iat, ix
# https://www.jianshu.com/p/199a653e9668
for i in range(df.shape[0]):
    df.iat[i, i] = 0
    df.iat[df.shape[0]-i-1, i] = 0
df
```

### 67. This is a question related to understanding of grouped dataframe. From df_grouped, get the group belonging to 'apple' as a dataframe.

```python
df = pd.DataFrame({'col1': ['apple', 'banana', 'orange'] * 3,
                   'col2': np.random.rand(9),
                   'col3': np.random.randint(0, 15, 9)})

df_grouped = df.groupby(['col1'])
```

```python
['apple', 'banana', 'orange'] * 3
```

```python
# Solution 1
df_grouped.get_group('apple')
```

```python
# Solution 2
for i, dff in df_grouped:
    if i == 'apple':
        print(dff)
```

### 68. In df, find the second largest value of 'taste' for 'banana'

```python
df = pd.DataFrame({'fruit': ['apple', 'banana', 'orange'] * 3,
                   'taste': np.random.rand(9),
                   'price': np.random.randint(0, 15, 9)})
df
```

```python
# Solution
df_grpd = df['taste'].groupby(df.fruit)
df_grpd.get_group('banana').sort_values().iloc[-2]
```

### 69. In df, Compute the mean price of every fruit, while keeping the fruit as another column instead of an index.

```python
df = pd.DataFrame({'fruit': ['apple', 'banana', 'orange'] * 3,
                   'rating': np.random.rand(9),
                   'price': np.random.randint(0, 15, 9)})
```

```python
# Solution
out = df.groupby('fruit', as_index=False)['price'].mean()
out
```

### 70. Join dataframes df1 and df2 by ‘fruit-pazham’ and ‘weight-kilo’.

```python
df1 = pd.DataFrame({'fruit': ['apple', 'banana', 'orange'] * 3,
                    'weight': ['high', 'medium', 'low'] * 3,
                    'price': np.random.randint(0, 15, 9)})

df2 = pd.DataFrame({'pazham': ['apple', 'orange', 'pine'] * 2,
                    'kilo': ['high', 'low'] * 3,
                    'price': np.random.randint(0, 15, 6)})
```

```python
df1
```

```python
df2
```

```python
# Solution
pd.merge(df1, df2, how='inner', left_on=['fruit', 'weight'], right_on=['pazham', 'kilo'], suffixes=['_left', '_right'])
```
